##### Statistical approach

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("C:/Users/hp/Downloads/simmons.xlsx")
df.head()
```

```python
plt.scatter(df.spending,df.coupon,color="red",marker="+")
```

```python
sns.pairplot(df,hue='coupon')
```

```python
x=df[['spending','card']]
y=df['coupon']
```

```python
import statsmodels.api as sm
x1=sm.add_constant(x)
logit_r=sm.Logit(y,x1)
result=logit_r.fit()
```

```python
result.summary()
```
